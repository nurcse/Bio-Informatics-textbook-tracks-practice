# BA1C - Find the Reverse Complement of a String

```python
def complement(seq):
    l = len(seq)
    #print("seq : " + seq)
    complement_seq = ""
    for i in range(0, l):
        if(seq[i]=="T"):
            complement_seq += "A"
        if(seq[i] == "A"):
            complement_seq += "T"
        if(seq[i] == "G"):
            complement_seq += "C"
        if(seq[i] == "C"):
            complement_seq += "G"
    return complement_seq

def rev_complement(seq):
    #print("SEQ : ",seq)
    complement_seq = complement(seq)
    #print("com_seq : ", complement_seq , "\n")
    return complement_seq[::-1]

print("rev_com : ", rev_complement("ATGGCGACGTTTTCTCTGCGGTGATCCAGTCCGGCGAATCGTGTTCGTGGTAATACTGCACTATATGACCCGAACGATAATGGATAAACACGGATGGCAACCTAGTCTATCTCTACCATGCTCCACAGCCAACTAACTTCACCCTCTAACTCTCGTGAGAGGGGAGCCGACGGCTTAGCAGTCCGAAGCGTACCCTATGCGAAAGAGATCGGTCGCGTGTATCAGATCATTCACAGATCCGCACAACTACGATTCGAAGGTACTGAACGACTCCACGCCCTTCCACCCCCGATGTGAAAGGCTAAAGCGTGCTCTCCTCGCTGCTAATCAAGGTGAGACGGCAGGTGAGCGTCGCTGCACCTCAGAACTGTAACGTTAGCGCATTACCTAATACGAGCAGGGTAGAGGATAATAGGGGAATAGGTGCTAAGGCGCAATGTGTCGGCGGTGGCTTCGATAGTCTAGGTCACTTAATCGCATAGGGTGGACGTGTGCCATGAACCTCTAGAGCTGGGGCTAGGCGCAACACCTACAGTTAATCAGACCCAGAACAGATATTGCTCGCAGTCAAGGTGACGTTACGAATGGACACTATAATTCCTCGAAAGGAGAATATACTCTAAAGGAATCGCGCTCACAGCCGCTTAAAGTACTCACATTAAGGCCATTTGCAACTATAAAGCGGCTGGGGTGCCACTTTTATGGAATAATGGAAAGCGCGTTCTTGGTAGCAATTGCCGACCAGGGGGCTCCAATACGTAACCATTGCCTTACACTAAGCTCCACTACGGGTGGTTAAATACGCTGTACCTACTCTCAACACTACTCCAAAACGTTAAGCACGAATACCTCGTTCTGCATTGAGCATCATTACAACCCCGTCACATTGGAGTATTTCTATAAATATAAATTTCCCATTACGGATGAAGTTCCATTAGTAGAGATACACTCCATCTAGTTATGAGGCAGATGATTGGCGCAATCCTGCACCCACCCAGTACTTCCACTGCTCGGTGCACTGCAATCTCCTCAGTTTCCGTGTTCATAGGCAGCTTATACTTATCGAAACTCAGCGACCGCGTTGTCTGGGGCGAAGTTAGATTGTCCCGTATTTCCTTCAAGATTCTAAATTCGTACGATCCCCACCATCACAGACATCTACGGTCTACCGCGCTTAACACTCCCCCAGTCCGAAAGGTGCCTTTGTAAAATCTACGTACGAGTTTCGGGAGCTTGCGAACTCTACTAGAGTTGGCTTATCGCCTACCCTCCTGAGGTTGAATTCCCTTAGGACGATCGAAGCGTAGATATCCTCAAGGACCGAGATTCAGTGTTGGTCACAGTTTCGCCCCGCTTCGGATCCAGCTCCGTTATGATCAATTTGTGTAAGGCCAGCCGGGCGACACAAGTGGCTAGATTCACCGCCCGGATCTAAATAGTAAATAGACTGAATACCCACAGAGTTCGTCATCCATTAGGCAGGGGCTGGAGTACCACATACAGTAGTAATTGGTGACCCAGAGCGCGTCGCTTCTTTCCTCAAGACTTAGTTTAAGGGGAACTTTTTCGCAAACGATGTTGTTGGATGCGGATCGATATAAGAGCGTTAATTATTTATGGGCGACGTGTCGCTGCGCTGTTCACGAAGGAGTGATCCCTTGTATTCCGAGGTGACCGTAATTTCTGGGGCGCTTTGATGCTGCACTGGACCCCATTGCAGTAGATTCTTGACCTACTGTTCAGTAGCAGACATTTGGAGTGACGTTGTCGTATAGGGACTTGAGACGGTTAGATTACAGGCGTTATGAGTAATGGGCGATTGATACACAATTTTGGGCCGACTCACGGTTCCGGTATAGAGCGGCGGTCTGTTATGCATATCATGTTAATTCAATTTAAGACGGAAATAACTGTGCGGAATCTGTGGATTTGCCACATATTAGCCATTCGGGCCGATTAATCCATAACTTTCCTTAGGATTGATTCACTACCGTTGTGATAGAAGAGTTTCCCTTGTATAAGAAGACTAAGTCTTGAATAATTAAATCTAAGAGAGGACTTTTTGGCCTTTTAAGTTGCTCAGGAGACACTGGTTTGTCGGAACCGTTTGGTATGATACCCATCAGGGCAAATTTAATGTACAACATAAGGAATCACGTCCCCTCCGAGTCATCACGTGTCCGATCAACCGGATAAGCGGGTATGAGCAAAGCCTGCCTAGTAGCCCAATTTCATTCACTACAACAACACCACTGGCGGTTCATTTATAGTTTGGCCGAACCCTGACAGCCTCACTCAACAACACGAAGCCGTACCGGATTGGCTGTGCAATCCAGTAAGGAGGAGCATACTTAGACGCTATAGTATTACTCGCAACGACCATTTATGGAATTCAATTTACCATAGCTACTAGAGCCCAGGGCTTCTACTAGGTCAGGCACGGGGTCCGGAAGCATCAGGAGCTCAGTAAACATTCCTACCCCTGTAGACAGCAATATTGTTATAACTCACGGCTCCTCGATGTGGGATTCGCTAGACCCCCGATGTGCTTCCAGTCAAACCTTTGTCGCTGAGTTATGCACACGACCTATTCCAAGAGATCAGATTAAGCAGGTATGTATTAAATCAGCAGAAGTCCATATCACAAAGCAGATATGAGAAAGACAGTCCGACCTATACCGCTACGTGACCAGCACGTCAGCCGAAGGGAATGGGATGATCGCGTGATCACAAAAAAGTGTGTTTGTCTGTAGGGTTGATCTATGAAGGCCTTTGCATATAGTTTCATCTCCAAGGCGTCTAGGAGGGCTCGAGCGGTCGGATAACGTAAATCCTCGCGGGCAGGAAATATAAACATATATCGAGTGTCTATACACGAGCTCACCTTGTAGGCTTCGAAAAGTTGGCAGTCCTTATGATCATGCTTACAGCCACATCCGCCGTATAAAATATTCCTGCACTAGTATGTACTGGCTAACAAATATAGATGGCAAAGCGGAAGTACCGAAAAGAGACGCCTAACGGCAGCTGGGGGACCAGTCGCTCGTGCCGGTTACAGACGGGCGTGAGGGTTTCACAATCCCCTAAACAAGCATAAGCAGAAGCGGGGAGTGTGAGACAGTCTCATCAGTGAAAGTGGGAATGAGCTACTTGTATTGTATTAGCCTGGTTGGCAAGACTTCAAACACGCGTAACGTACGTAGCTATGCGGTGCATCACTAAAAAGACTTATACCAACTCGCCGGAGGCCGTCTTCGCCCATCTTTACTTTCCTTCTAGGATTTAGCCGTCGGGTTTCCTATGAGAAACAGCTGCGAGAAGCTGTCTTCTACATGCTGGAATGGTGGACCTGCATCTGATCGGTATATAGCTACCAGGCGTGCGGCTCCTGGCTAGAATTAGAGTCAATTGGTGAGTTGCCAAGACATTGGTCGCCCGCCTCCCTTCATTATACGGTAAATCAAAAAATACTTCTTACCTTAATGATTGTACATCGGTCTGGTTTCGATGATCACTACAACATTTAATTGGTATCTCTGTGAGCATCTTCAGCCCCATAAAGCAGAACTAGGAGGGAATTAATAAGAAGTAGGGAAAAGCGTCTTACTGACGTTGTTGCAGGCACAAGTAAATAGGATTAACAGGTTAGTGCGGCCATTAAATGAAGTAATCTTTTTTGTACCTTCCTATTGCTTATGGATCGTCTGTCGGTAACACAGAAAAGCGCGCTTCTGAAGGCCGAGACTTTCTCGGTATATTGGAAAGTCGCACGATGCTATAGTGCATACGCTAAGGTCTCCGAGTTGCAGGTTTATATGAACCGTTTGCACGTGGTGCTAAACCGTAACTCTCCTTTCTAGATATACCTCTACCAACTTAGCAGCATGAGAATAAAACAGAATAAGTGTAAGCCACGGTTGTGTGGGAGAAGAGCGAGTATAGGACTAGCATAGCACTTCGCTTTCCGCCATGATAAGTGGATTAGATCGCTCCAGCATTATAGTCTACGGGATGCCGTCTGATTCTATCCTCCCTGTCACGCGACGACATCTCTCCGGCTTTCAGGCGTATTTCCACTGGCTGCTTCGCCGGTTTCGGACGTCCGTCCAGCATAAACCGGCACATGTTCGGCCATATTCCTAAGGTAGACTGAAAGCGCCGACTCTGACCCTAATCGTAGTGGCCCATCGACCCAGTGGTTGGGAATCAGACTTAGAAGTAACAGGTTTGTTCGAGCCATAGCGGAGTACTCTGTTGCTCCCCTGAGCGTGGATTCGCTACAAACTAGGCGACATTGGTTCGGATCAAGGTCGGTAACACAATATCTGTACTGTGGTGACTGTATACACACGCTAGGCACGGCTTTAAGTTTTTGCGGAATTGCACCCTTATTGCACCTGCGAGTCTATGTCATGTAGTAACTGATGCTTTAACTAGAAAGGCGCGCTAACTTCAGGACGTGCTAGTCATGGATTGTAGTTGGTAGTGTCGGCGTTAGGTTGGGTTTCACGTAGGATTGGAAGTCTGACATGCCTACGGGAACACTGTTCCGTCAGGGCAGCCGTCTGCCTTACGCGCACACCCACACTTAGTGCTATGGAACTATCCGCACCAGTGCCCTAGGGTTTAAGTCAGAAACCATCGAACACAGCCGAGGTCACGTAGTAGGCCCTGACCTTCGTTTCTTGTGCGAAGTCAGGCGGGCCGCTAATGGGGGGACACACCGCGCGTTGTTACTTGTCGCCTAAACGCAGTTATCTAGCCTGAGCCCAACGTGCATTCAATAGTAGAGGCCTTGTACGTACCTACCGATCTCTTATATCATGTGCGTTCACACTCGTGCGCATAACTAGGAGTGGTGGAAGGCATAAAAAGAAGCAGTTGCACAGAACTGAATAAGACCGTTTTCAAAGTCCCATTAGCCTCATCACTAGATCAGGGGCACAGAAAGCCATCCGCTGCGCGACTGTTATTTCCGTTGCTATGTTAGAAAGCAATCCGCTATATACCGTCCGTCATATTACAGACATGGACTCAGTCAAGAGGGCAACCTATGACTAGCTGGTTCGATATCCAGGGTGGCGGTGAATACCGAGCCGTTACGACATTTGCTTGTGGGGAGGATACCGACTCGGGGACTGACCAAACACCATTGTTACAATATTCAGTGTCGGTCTCGATGCGTCTCGCCGGACCTTAAACAGAGGTCCTAGTTGCAGACATAAAGGTCAACCGACCAAAGTGATAGAGAGCCCACGCATGATTGAGCCCCATTCGCCAGTTCATTGCATGGTCGAGTTCGTTACTGCTCGTTCCCCGTTGTGCAGGACACATTTGGGTCCTGCGCACTAAGACTGTCAGGCTTAATGCAGCTCAGGAAGGTAGTCAAGCGCACCGTTTTGTCCAACCCTAAGCCGCTCCCCGCTCCGGCAGACAAGGAGGTGAACTTGTTCGTTGCATATCCTCGACCAATGTTTAGTCCACGTGTCGTTCCATCCCGCCTTTTCCAAACTTTGTGGATTAATATTCCATGTAAGCGTCGTGCTCTACTTGTACGGATGTTGACGGGGAACCGACAACGTCTATCCCGGAGACTTGAGCTGTACGGTAGACACTTCAATAGAATATCTCCACCATCACCCGCGCTGGGGCTCAAGCCTCAGCAACATTATCGCTAATATGGGAGACGTGAGCTGCGAGTCCCGGTTCGATTCTAACCTTGACAGCTACTACCGAAGTGGTTACTAGCACGTGAGGTATGGAGCAGCGGTTACTGCGCCGTTTACTATGATGAGTGCAAGCACGGGGGGTGCACGTCTTTGTTATCTGGGTGGTTTAGGATGCAACCTCGCGATATGAGGTTGTGCTATACCATCAATAAACCGAGGTCGCTTAGATGATACCATCATTCCCAGGTAACTTCTGCGCATACGAAATGAGAAATTGGACGTTAGGCGGGATCAGGATTATCACCTATTATCCGATGTCGACCTGTCGTTTTAAGGAAATAGAACGGCTCGCACACGCATAGGAGCCGCACAGAGCCCCAGACTTCCTTATCGCTGTTGTTTCTCGACTCGGTAGTGTCGTGTGGTTCCCCCACTGGGCAGATCGTGTTCACCAGTTATCCCGAACGGGTATGGCTAAAAGATGTAATGCGAATTTGCTGTTGGGCCATTACCGGGTGACGGTAGCTCGGCAAAGCAAAGACGCTCGGGTAAATAATCGAAGGCAGCCAGACGCTTCACTGCGGTCGGAGATACATCAGTTTATAAATTATATGCGCGCAAGTTTACCTGTATCTGCTCCCTTTTGTCGGTCTTCAGTTACTCGTGTGATTTACGTCTTTAAGAGAATACGCCTCCTATATGTGTTGGGCTGGGCGGACCCCTATCTATTGCCCCCGGTGTCAGATCGATGGGAGGTACAGGAGAAAGGACATTCGGTTATTACCAGCCAACTATTGGTTCGAGGCATCCCTACAGGTGGTAACGTAAGACATCAGTAGTAGGATGGCAGCACTACGACTATGCAAGTGAGTCCTCAATGTTGTACTTCTGACGGTATCCTCCGGCCAGCTGCGATCCACTTGAAAAAGGACAGCCTGGTCCGGGCCAAGCTCGGGCCAGATGGACGGGGCGGCCTGACTACTTGGAAAGGTAGATCTGAACTGGATAGTGAAACTGCAATAGAGCCCGTCGTGTCATCGGAGGACAGTACCTCAAGCCAACCCGTAGGCTGTAAGCTCAGGCAGCGACCGAAGCTGCGCGTATCTCCCTAATCCTACCTTGGATATGGCACGATAGCTACTCCCATAAATGTCAATCTGTCGGAGGCCGGAAATAGAACAAGGATAATAGAATGATTACCGGTTACACATCTGATCGGCCGAACGCGGCCCTGTACACCAGACCAGTTCTTTTGTTATCTGTCCCCGGCCATGACCTTAGCTTTATGTAGCAAGTGCGTTTGCCAATCACGAGACTTGGGCATGTAGCTTGCAACGTGGCACAACACTGCAGACGCCAGATCATTCACACGCATGTGCGACGAAATACCTGTAATCTCCAAGACTCTGGAGGGGACTCAGGAGCTTTTCTTAGTTAGCGCATTCGCAAGCTCGCTAGGTAGAGTACTGAAAATGCTGGAGTTTGAATAGGTGTTCAAAGTACCGCAGTTTTCGCGAGTGACTGAGGAGACCCGGTAGTTTTAACCCCGGGGAGTGCTCTGCCTTAACGGGAACCTGACAGCCGTGCGCGCCATCGAACATATAACATGCACGAACATAGTAATACTTAGTTGCGGTAGTCTTCGAAGCCCCCTATAAGTGAGGCTTATCCATTGATACCCACTACTCATGAGCTGCATACGCGGTGTCTGCTCGTCCAATAACGATTATTGGTTGTGCTGGTCCTGCTAGCATCTCAGGCTACCATTCCCAGAAATAGCCTCCAGTGGTGTGTGTGATGACTATTAACCAGTAGCAAGGAGTGTTATGGCTGCCCATTTCACGGGTCTTCCCACGGCTTACGATAAGTGGTTAAGGGCTGTTGCTAATCTTAAAAAATCGAGAATATCAGACCCTCGGTCCCCCCCCGATGTTTTATTACCCGCCAACAGCTAGTGGTCGAGTCCGCCTAGTCTAGCGTGGCCTGTATTGCACGTTAAACCGGGACTACTCTGTGACGTCCTCATACCGAATGAAGCGTGAAATTCTGTACCCGCTTATCCTGGCTAAGTATGGGTTGAAGCGTTTGCACTACGAGCGGGGTGGAGTTGTGGGGTGATCTTGCGAACACCACAACCTTAGCCGTAGCAAGGAATAAAAACGGAAAAACATACATAAGAATGTACAGGAGGCATCATCTGAAAATCTTAGGTCTGTCAAAGACAAGGCACTTTTAATGAAGAGTTACCATAGAAAGCTACCCTATCCATCTCAGGCAATGAGTTTCATTGTTTTGCTTTCCTTTAACTCTAAACCGAGAGTTCTGGATTGGGTATTCCACATGAGATGAATATCTACGGCATTGTGGCGGCTGTAACATCGAGGCAGTACTTCGGTTCCTAGGCTAGCACGGTTTGCGTTTGGACGGTGACCTTGTTACTTTAACGCTCCATCCCCAGATACATCTAAAGCGATAACCTTAAGAAGTATTGTTCAATAAAATTGTTGACCGACTGATAAAAGTGCCGCATATCTAATGTACCGCTGGCAACCGGATGCTTGGTTACAGTCAGGTGCCAACATTCTTCACCGAGCCCAGCTACCTGGGATATCTACGCTGACTTTGTGAGACTAGAGCTTGCGCATTCGGATTCAGTACAGCGGCGCACGCTGTAACTTGGCGGAGACGCCCAGCCCTCCCGAGAGAGAAAGCGTACTTAGGATATGACGAAATGCATCGGCTAATTAATGTCGGTCAAGAATGTGTTACAAAAATTATCACATGGCCGATAAACAGAACAAACCTTGGCATTTCCGAAATCACTGGCGGTACCAGTGGGATCAGTTCTTAGGCGGTGGAGGATTTAATCCACTCATAGCCACCTTGGCTCCATCCGCCCCAGATCTTTAGTACTGCTCGACGTTAGTCATAACACCAACTGACAGTGCTGCGAAGATGGTGATGGAACGTTTTTCCCCACCTGGGTCCTCTTAGTAGCCCCTCTGACACTGGCTGGGATTAATTACGCTTTGATATTATAAGTACTGGCGGAGTTTAGGTACCCGTGCCGTTCCCCGGCTATGACGCGCTTGTTACATATTCGACAGCCCGGACGTGCACCAATCAATCCAGTGCGTCGCGTCCGTCAAGCGGGCGTGTCCTTTGCTGGATGAACTTTAGTCCTTCCGTGACGAGTACGATTATCTTTTGTGGGGGCGCCAATAAAAACAGCTTGCGGCAACTCTACCATGCTGAGTTATCTATGCTTAGTTTAGTCGCGCGTGCGTAGCCCCTTAACGGCGTAGCTGAAAGGGTACACAAGGGAGATTCATGCACCTTCCTCCGATCGAGTGCCCGCGTGGGCTAGGGCCTTCTGTCAAGGGGGCGTATTTGGGGCAATAGAGAGCGGAGTAAACTGGCATGATCGCCCAGCGGACAGACGGGTTAAACGAGATAACTGGTTCATTGGGAATAGCATTATGTCAGCTTTAGCGTGGCATGAGGATAAACTTATTCTTGGGTGAGCGTCTCGAGGCAGCGGTTCCCCCGCCCTGGAAGGGCGTAGTGGGCTCCCGCGACACTAGGTACGATCTGACATCTCTGCAGCTCTACAGGTTGTGTCTGTCCAGAAATAGACCCCGAGTGATGCAGACAGGTCCGTGTAGATGTATCCGACAATCGCGGCATGGGTAAATCGTTCATCAGAG"))
```

```python
def patternToNumber(dna):
    l = len(dna)
    res = 0
    dic = {'A':0, 'C':1, 'G':2, 'T':3}
    for i in range(0, l):
        res = res + ((4**(l-i-1)) * dic[dna[i]])
    return res

#print(patternToNumber("CTTCTCACGTACAACAAAATC"))
print(patternToNumber("CCATGGC"))
            
```

# BA1M - Implement NumberToPattern

```python
def numberToPattern(num, index):
    mod = 0
    divisor = 0
    res = []
    divisor = num//4
    mod = num%4
    print(divisor)
    res.append(mod)
    while(divisor != 0):
        mod = divisor%4
        divisor = divisor//4
        res.append(mod)
    res = res[::-1]
    dic = {0:'A', 1:'C', 2:'G', 3:'T'}
    result = ""
    if(index > len(res)):
        a = index - len(res)
        for i in range(0, a):
            result += 'A'
    for i in range(0, len(res)):
        result += dic[res[i]]
    return result

print(numberToPattern(6564, 9))
```
